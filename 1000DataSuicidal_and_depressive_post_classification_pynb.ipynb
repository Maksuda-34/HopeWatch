```python


def rem_all_special_characters(text):
    pattern = re.compile('[@#$%^&*()_+-={}\[\];:\'\"\<>,.///?`~।]', flags=re.I)
    return pattern.sub(r'', text)
def rem_special_characters(text):
    pattern = re.compile('[@#$%^&*()_+-={}\[\];:\'\"\<>.///`~࿐❛❛❜❜]', flags=re.I)
    return pattern.sub(r'', text)
def rem_non_bangla_characters(text):
    pattern = re.compile('[A-Z]', flags=re.I)
    return pattern.sub(r'', text)

def rem_emoticons(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'', text)


def remove_duplicate(dataframe, column_name):
    final_df = dataframe.drop_duplicates(subset=column_name)
    return final_df
```

```python
def remove_non_bangla_characters(text):
    return re.sub(r'[^\u0980-\u09FF .,?]+', '', text)

# Example usage:
a = "খারাপ? পুরুষের, সংস্পর্শ ছাড়া কোনো ভালো নারী কখনো খারাপ হয় না।"

a = remove_non_bangla_characters(a)
print(a)
#a = rem_special_characters(a)
#print(a)
```

```python
Data = pd.read_csv("TestData1000.csv")
Data.head(50)
```

```python
import time

def translate_and_clean(text_list):
    translated_text_list = []
    for text in text_list:
        try:
            if text is not None:  # Check if text is not None
                translated_text = translate_to_bangla(text)
                cleaned_text = remove_non_bangla_characters(translated_text)
                translated_text_list.append(cleaned_text)
            else:
                translated_text_list.append('')  # Return empty string for None values
        except Exception as e:
            print("Error processing text:", text)
            print("Error:", e)
            translated_text_list.append('')  # Append empty string in case of error
    return translated_text_list

# Assuming Data is your DataFrame and 'Post' is the column containing text to be translated
batch_size = 30  # Adjust batch size according to your needs
for i in range(0, len(Data), batch_size):
    Data.loc[i:i+batch_size, 'Post'] = translate_and_clean(Data.loc[i:i+batch_size, 'Post'].tolist())
```

```python
Data.head(60)
```

```python
Data = Data[['Post', 'Tag2']]
Data.head(30)
```

```python
#Convert 'Tag2' column to categorical
Data.Tag2 = pd.Categorical(Data.Tag2)
#Encode categorical values
Data['Tag2'] = Data['Tag2'].cat.codes
#Remove rows with missing values (NaNs)
Data = Data.dropna()
Data.head(50)
```

```python
# random_state=880 (Bangladesh country code)
from sklearn.model_selection import train_test_split
#splits the Data DataFrame into training and testing sets, with 80% of the data used for training and 20% of the data used for testing, using a specific random seed (880) for reproducibility.
train, test = train_test_split(Data,test_size=0.2, random_state=880)
```

```python

len(train.Tag2.unique())
```

```python
test.head(50)
```

```python

from simpletransformers.classification import ClassificationModel
```

**Bangla-Electra**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score , f1_score
```

```python
# set use_cuda=False on CPU-only platforms
# set use_cuda=False on CPU-only platforms
model = ClassificationModel('electra', 'monsoon-nlp/bangla-electra', num_labels=3, use_cuda=True, args={
    'reprocess_input_data': True,
    'use_cached_eval_features': False,
    'overwrite_output_dir': True,
    'num_train_epochs': 3,
    'silent': True
})
model.train_model(train)
```

```python
print(test.columns)
```

```python
# Make predictions on the test data
predictions, raw_outputs = model.predict(list(test['Post']))

# Convert 'Tag2' column to a list
true_labels = test['Tag2'].values.tolist()

# Calculate accuracy
accuracy = accuracy_score(true_labels, predictions)

# Calculate precision
precision = precision_score(true_labels, predictions, average='weighted')

# Calculate recall
recall = recall_score(true_labels, predictions, average='weighted')

# Calculate F1 score
f1 = f1_score(true_labels, predictions, average='weighted')

print("Accuracy:", accuracy * 100, "%")
print("Precision:", precision * 100, "%")
print("Recall:", recall * 100, "%")
print("F1 Score:", f1 * 100, "%")
```

**Bangla-Bert-Base**

```python
bert = ClassificationModel('bert', 'sagorsarker/bangla-bert-base', num_labels=3, use_cuda=True, args={
    'reprocess_input_data': True,
    'use_cached_eval_features': False,
    'overwrite_output_dir': True,
    'num_train_epochs': 3,
    'silent': True
})
bert.train_model(train)

```

```python
# Make predictions on the test data
predictionsBanglaBert, raw_outputs = bert.predict(list(test['Post']))

# Convert 'Tag2' column to a list
true_labels = test['Tag2'].values.tolist()

# Calculate accuracy
accuracyBanglaBert = accuracy_score(true_labels, predictionsBanglaBert)

# Calculate precision
precisionBanglaBert = precision_score(true_labels, predictionsBanglaBert, average='weighted')

# Calculate recall
recallBanglaBert = recall_score(true_labels, predictionsBanglaBert, average='weighted')

# Calculate F1 score
f1 = f1_score(true_labels, predictionsBanglaBert, average='weighted')

print("Accuracy with BanglaBert:", accuracyBanglaBert * 100, "%")
print("Precision with BanglaBert:", precisionBanglaBert * 100, "%")
print("Recall with BanglaBert:", recallBanglaBert * 100, "%")
print("F1 Score with BanglaBert:", f1 * 100, "%")
```
